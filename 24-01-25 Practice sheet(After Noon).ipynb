```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
categories = ['A','B','C','D']
val = [20,32,45,66]
plt.bar(categories,val,color = 'skyblue')
plt.title('Basic bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
plt.bar(categories,val,color = ['red','green','black','purple'],edgecolor='yellow')
```

```python
plt.bar(categories,val,color = ['red','green','black','purple'],edgecolor='yellow',width = 0.021)
```

```python
ages = [20,21,22,23,21,23,24,25,23,25,26,24,25,25,31,32,34,35,40,42,50,51,52,55]
b = [20,30,40,50]
plt.hist(ages,bins = b,edgecolor = 'black',color='green')
plt.title('Age distribution of customers')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
plt
```

```python
P = ['Python','Ruby','C++','java']
Pe = [40,30,20,10]
myexplode = [0.1,0,0,0]
color = ['gold','lightblue','lightgreen','pink']
plt.pie(Pe,labels = P,startangle = 90,explode = myexplode,shadow = True,colors = color,autopct = '%1.1f%%')
plt.show()
```
